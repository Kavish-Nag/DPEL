Data scaling is one of the most important data preprocessing step in machine learning.

- The most common techniques of feature scaling are:
  - Normalization
  - Standardization
- Normalization is used when we want to bound our values between two numbers,typically, between [0,1] or [-1,1].
- Standardization aims for a mean of 0 and a standard deviation of 1.

Normalization or min-max scaling is used to transform features to be on a similar scale. The new point is calculated as:

Xnew =(X-Xmin)/(Xmax-Xmin)

For, standardization or z score normalization is the transformation of features by subtracting from mean and divind by SD. This si often called as Z-score.

Xnew=(X-Xmean)/Xstd_deviation

```python
#import libraries
import pandas as pd
import numpy as np
```

# 1. **Min_max scaler**

Import min-max scaler (normalization_scaler)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Define data
data=pd.DataFrame(data=[[5,5],[10,2],[2,0],[0,10]],columns=['X','Y'])
data
```

```python
#plot scatter plot
import matplotlib.pyplot as plt
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c="r") #axis vertical line
plt.axhline(y=0,c="y") #axis horizontal line
plt.show()
```

```python
#define,fit and transform min-max scaler
scaler=MinMaxScaler()  #y=(X-Xmin)/(Xmax-Xmin)
min_max_scaled=scaler.fit_transform(data)
min_max_scaled #prints in numpy array type form
```

```python
min_max_scaled=pd.DataFrame(min_max_scaled,columns=data.columns) #or we can write it as columns=['X','Y'] also
min_max_scaled #prints in datafram formm
```

```python
#Plot both min_max_scaler
plt.scatter(min_max_scaled.X,min_max_scaled.Y)
plt.axvline(x=0,c="b") #axis vertical line
plt.axhline(y=0,c="c") #axis horizontal line
plt.show()
```

#SUBPLOTS

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) #1row,2column , column 1 position (index of figure)
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c="r") #axis vertical line
plt.axhline(y=0,c="y") #axis horizontal line

plt.subplot(1,2,2) #1row,2column , column 2 position (index of figure)
plt.scatter(min_max_scaled.X,min_max_scaled.Y)
plt.show()
#if index of figure not given, it will print one below the other
```

```python
#Import standard scaler (standardization_scaler)
from sklearn.preprocessing import StandardScaler
#define data
data=pd.DataFrame(data=[[5,5],[10,2],[2,0],[0,10]],columns=['X','Y'])
data
```

```python
scaler1=StandardScaler() #Create a StandardScaler object for z-score normalization
#Fit to 'data' and transform it to have mean=0, std=1
std_scaled=scaler1.fit_transform(data)
std_scaled
```

```python
#Convert standardized array back to DataFrame with original column names
std_scaled=pd.DataFrame(std_scaled,columns=data.columns)
std_scaled
```

```python
#Plot standard scaler
plt.figure(figsize=(5,5))
#Scatter plot of standardized X vs Y values
plt.scatter(std_scaled.X,std_scaled.Y)
plt.axvline(x=0,c="b") #axis vertical line
plt.axhline(y=0,c="y") #axis horizontal line
plt.show()
```

```python
#Summary statistics of standardized data
std_scaled.describe()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) #1row,2column , column 1 position (index of figure)
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c="r") #axis vertical line
plt.axhline(y=0,c="y",label="y=0") #axis horizontal line

plt.subplot(1,2,2) #1row,2column , column 2 position (index of figure)
plt.scatter(std_scaled.X,std_scaled.Y)
plt.axvline(x=0,c="r") #axis vertical line
plt.axhline(y=0,c="y") #axis horizontal line
plt.show()
#if index of figure not given, it will print one below the other
```

```python
#Plot bothn min-max and std scaler
plt.scatter(std_scaled.X,std_scaled.Y , marker="^")
plt.scatter(min_max_scaled.X,min_max_scaled.Y, marker="o")
#Add vertical and horizontal reference lines
plt.axvline(x=0,c="b")
plt.axhline(y=0,c="y")
plt.show()
```
