```python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(1, 3, figsize=(12, 4)) #3 subplots arranged in 1 row
x = np.arange(1, 11)
# Plot cubic curve with default grid
axes[0].plot(x, x**3, 'g', lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
#exponential curve with customized grid
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls='--', lw=0.25)
axes[1].set_title('custom grid')

axes[2].plot(x, x)
axes[2].set_title('no grid')
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure and an axes object
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# Enable grid lines
ax.grid("True")

# Set the color and thickness of the spines (borders) of the plot
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2) # Make left spine thicker
ax.spines['right'].set_color('black')
ax.spines['top'].set_color('green')

# Plot a simple line
ax.plot([5, 6, 7, 8, 1])

# Display the plot
plt.show()
```

```python
#using iris dataset
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# Load the built-in iris dataset from seaborn
iris = sns.load_dataset('iris')
# Display the first 5 rows of the dataframe
iris.head()
```

```python
#just the grid
     # load iris dataset
sns.pairplot(data = iris)            # create grid of scatterplots + histograms

#it creates a grid of scatterplots and histograms for every pair of numerical features in the Iris dataset.
```

```python
sns.pairplot(iris, hue='species', palette = 'rainbow')
#mentioning hue colors the plots based on the species column
```

pairgrid mechanism

```python
sns.PairGrid(data = iris)
#base framework for creating a grid of subplots
```

```python
#then you map to the grid, pairgrid
g=sns.PairGrid(iris)
g.map(plt.scatter) # fill all cells with scatter plots
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist) #Histogram (one feature at a time)
g.map_upper(plt.scatter) #scatter plots (pairwise relationships)
g.map_lower(sns.kdeplot) #KDE plots (smooth density estimates)
```

```python
#using tips dataset
tips = sns.load_dataset('tips') #inbuilt tips dataset
tips.head()
```

```python
#just the grid
g = sns.FacetGrid(tips, col = 'time', row = 'smoker')
#splits the plots into columns based on the time (Lunch / Dinner)
#splits into rows based on smoker
```

```python
#the total bills are plotted as histogram across the grids we created using 'map'
#question
g = sns.FacetGrid(tips, col = 'time', row = 'smoker') #FacetGrid is used to create a grid of subplots
g.map(plt.hist, 'total_bill') #Plots histogram of total_bill
```

```python
#over here we have defined hue as sex
#and also plotted a scater plot where x axis is total bill and y axis is tip
g = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex', palette='coolwarm') #hue divides based on the species column
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

Joint grid

```python
# JointGrid creates a plot with a central bivariate plot and marginal distributions on top and right
g = sns.JointGrid(x='total_bill', y='tip', data=tips)
```

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tips) # create a JointGrid comparing x and y axis
g = g.plot(sns.regplot, sns.distplot)
```

```python
#implementation of matplotlib function
#plotting with and without grids
import matplotlib.pyplot as plt
import numpy as np
#dummy data
#by default, num=50 if not specified
x1 = np.linspace(0.0,5.0) #generate 50 points evenly spaced between 0 and 5 (start,stop,num)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) # damped cosine function (decays with time)
#creates two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
#plot without grid
ax1.plot(x1, y1)
ax1.set_title("Plot without Grid")
#plot with grid
ax2.plot(x1, y1)
ax2.set_title("Plot with Grid")
#draw gridlines
ax2.grid(True)
plt.show()
```

```python
#set line properties for the grid
#draw gridlines using extra line properties such as color, linestyle and linewidth
#implementation of matplotlib function
import matplotlib.pyplot as plt
import numpy as np
#dummy data
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
#set graph color
plt.plot(x, y, 'green')
#to set title
plt.title("Plot with linewidth and linestyle")
#draw gridlines of grey color using given linewidth and linestyle
plt.grid(color='grey', linewidth = '0.4', linestyle=':')
plt.show()
```
