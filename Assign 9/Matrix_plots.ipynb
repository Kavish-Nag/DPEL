#Matrix Plots
They allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data.

```python
import seaborn as sns
%matplotlib inline
```

```python
#importing the modules
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#generating 2-D 10x10 matrix of random numbers from 1 to 100
data=np.random.randint(low=1,high=100,size=(10,10))
print("The data to be plotted:\n")
print(data)
#plotting the heatmap
hm=sns.heatmap(data=data)
#displaying the heatmap
plt.show()
```

Displaying the cell values

```python
data=np.random.randint(low=1,high=100,size=(10,10)) #randomly generate a new set of numbers
annot=True #already making it True in a variable
hm=sns.heatmap(data=data,annot=annot)
plt.show() #display the plotted heatmap
```

```python
data=np.random.randint(low=1,high=100,size=(10,10)) #randomly generate a new set of numbers
cmap="tab20" #already defining the color mapping
hm=sns.heatmap(data=data,cmap=cmap)
plt.show() #display the plotted heatmap
```

```python
data=np.random.randint(low=1,high=100,size=(10,10)) #randomly generate a new set of numbers
#setting the parameter values
linewidths = 2 #note: it's linewidths and not linewidth
linecolor = "y"
hm=sns.heatmap(data=data,linewidths=linewidths, linecolor= linecolor)
plt.show() #display the plotted heatmap
```

#Hiding the colorbar


```python
data=np.random.randint(low=1,high=100,size=(10,10)) #randomly generate a new set of numbers
cbar=False #used to hide the colorbar on right side on the graph
hm=sns.heatmap(data=data,cbar=cbar)
plt.show() #display the plotted heatmap
```

```python
data=np.random.randint(low=1,high=100,size=(10,10)) #randomly generate a new set of numbers
#setting the parameter values
xticklabels = False
yticklabels = False
hm=sns.heatmap(data=data,xticklabels=xticklabels, yticklabels=yticklabels)
plt.show() #display the plotted heatmap
```
