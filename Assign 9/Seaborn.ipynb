```python
#import seaborn
import seaborn as sns

#apply the default theme
sns.set_theme()

#load the iris dataset
tips = sns.load_dataset("tips")
```

```python
# Load the tips dataset again (already loaded above)
tips = sns.load_dataset("tips")
# Display the first few rows of the dataframe
tips.head()
```

```python
# Create a bar plot of tip amount by sex
sns.barplot(x='sex', y='tip', data=tips)
```

```python
import numpy as np
# Create a count plot of the number of tips by day
sns.countplot(x='day',data=tips)
```

```python
# Create a box plot of total bill by day
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
#Can do entire dataframe with orient='h'
# Create a horizontal box plot for the entire dataframe
sns.boxplot(data=tips,palette='rainbow',orient='h')  #if orient not mentioned, always vertically plotted
```

```python
# Create a box plot of total bill by day, separated by sex using hue
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips,palette='coolwarm')
#hue gives different colors with unique values of that category  mentioned
#how much max money is paid and by male or female?
```

#violin plot
A violin plot plays a similar role as box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical data.

```python
# Create a violin plot of total bill by day
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow') #violin plot, palette can be any rainbow/coolwarm or others.
#the white line in plot shows the mean value.
```

```python
# Create a violin plot of total bill by day, separated by sex using hue
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1')
```

```python
# Create a distribution plot of total bill, separated by time (Lunch/Dinner)
sns.displot(data=tips,x='total_bill',col='time',kde=True) # overlay a Kernel Density Estimate (smooth curve) on top of the histogram
```

```python
# Create a distribution plot of total bill (deprecated function)
sns.distplot(tips['total_bill']) #dist plot is in older version of sns, replaced by displot
```

```python
# Create a distribution plot of total bill with 5 bins and KDE
sns.displot(tips['total_bill'],kde=True,bins=5)
```

```python
# Create a distribution plot of total bill with 5 bins and KDE (same as above)
sns.displot(data=tips, x="total_bill", kde=True, bins=5)
```

```python
# Create a distribution plot of total bill with 30 bins and KDE
sns.displot(tips, x="total_bill", kde=True, bins=30)
```

#Joint Plot
jointplot() allows you to basically match up two displots for bivariate data. With your choice of what kind parameter to compare with.

```python
# Create a joint plot of total bill and tip with scatter plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') #shows histogram along with scatter plot
#outside graph will always be histogram
```

```python
# Create a joint plot of total bill and tip with hexbin plot and red color
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',color='red') #
```

```python
# Create a joint plot of total bill and tip with regression line
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') #reg is regression line ( y = mx+c )
```

#Pair plot
It creates a grid of plots showing pairwise relationships between all numerical columns in a dataset.

```python
# Create a pair plot for the entire dataframe
sns.pairplot(tips)
```

```python
# Create a pair plot with hue based on sex and 'coolwarm' palette
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

#RUGPLOT
They just draw a dash mark for every pt on a univariate distribution. They are building block of a KDE plot.

```python
# Create a rug plot for the 'day' column
sns.rugplot(tips['day'])
```

```python
# Create a KDE plot and rug plot for the 'tip' column
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

```python
# Create a KDE plot with red color and rug plot for the 'total_bill' column
sns.kdeplot(tips['total_bill'],color='red')
sns.rugplot(tips['total_bill'])
```

#Swarm plot
It's a categorical scatterplot.It shows all individual data points of a variable, spaced out along the categorical axis so they don’t overlap.

```python
# Create a swarm plot of age by class with 'Set2' palette
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

#Heat map

```python
# Calculate the correlation matrix for the numeric columns of the tips dataframe
tips.corr(numeric_only=True) #Pearson's coefficient
```

```python
# Create a heatmap of the correlation matrix
sns.heatmap(tips.corr(numeric_only=True))
#darker the color, lesser is the value
#lighter the color, higher the value
```

```python
# Create a heatmap of the correlation matrix with 'coolwarm' cmap and annotations
sns.heatmap(tips.corr(numeric_only=True),cmap='coolwarm',annot=True)
#if we dont use cmap='coolwarm' , then it will use default colors - shown earlier
#annot (annotation) values the value as well of those colors.
```
