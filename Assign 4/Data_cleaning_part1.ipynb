# **How to drop or replace null values?**

```python
#Actual dataset
df
```

```python
help(df.dropna) #to get deatils/documentation of that particular cmd
```

```python
#Missing value count along columns
df.isna().sum()
```

```python
df.isna().sum(axis=1)
#counts the null values along the rows
```

```python
#missing value count of complete dataframe
df.isna().sum().sum()
```

```python
#drops Null values
df1=df.dropna(axis=0) #0 is row , 1 is column
df1
```

```python
df1.isna().sum().sum()
#counts total null values present in code
```

```python
df2=df.dropna(axis=1) #drops null values about column
df2
```

**#threshint , optional**
# Require that many non-NA values. Cannot be combined with how.

```python
#threshint , optional
# thresh=2 means: keep rows that have at least 2 non-NA (not missing) values
# If a row has fewer than 2 real values, it will be dropped
# Require that many non-NA values. Cannot be combined with how.
df4=df.dropna(thresh=2)
df4
```

```python
#fill Null values with 10
df.fillna(10)
```

```python
df['age'] #Displaying just the age column
```

```python
df['age']=df['age'].fillna(50) #Fill all missing values in age columns with 50
df
```

```python
df['first_name']=df['first_name'].fillna('John') #filling missing values in name field
df
```

```python
df.info() #gives datatype and other info about all columns
```

```python
df['last_name']=df['last_name'].fillna("Doe") #filling null values in last name
df
```

```python
df['pre_movie_score']
```

```python
df['pre_movie_score'].mean() #mean of all existing values of pre movie scores
```

```python
#fill missing values of pre movie score with mean of the existing values
df['pre_movie_score']=df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
df
```

```python
df['age']=df['age'].fillna(df['age'].median()) #fill null values of age by their median
df
```

```python
df['post_movie_score']=df['post_movie_score'].fillna(df['post_movie_score'].mean()) #fill null values of post movie score by mean of post movie score
df
```

```python
df['Gender']=df['Gender'].fillna("m") #filling missing values in gender
df
```
