# **KNN imputer for quantitative or numerical data**

```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.impute import KNNImpute #Imported the KNNImputer
```

```python
dict={'Maths':[85,91,np.nan,91], 'Chemistry':[72,75,74,np.nan],'Physics':[np.nan,68,82,81],'Biology':[81,81,72,np.nan]} #Created a dictionary with marks in different subjects
```

```python
Before_imputation=pd.DataFrame(dict) #converting dict into df
Before_imputation
```

```python
# Creating a KNN Imputer that will look at 10 nearest neighbors to fill missing values
imputer = KNNImputer(n_neighbors=10)

# Applying the KNN imputer to the dataset with missing values
# It will fill in the missing data based on similar rows (neighbors)
After_imputation = imputer.fit_transform(Before_imputation)

# Displaying the data after imputation (missing values are now filled)
After_imputation
```

# **Convert the array into a DataFrame**

```python
Before_imputation.columns #columns of the df
```

```python
# Converting the imputed NumPy array back into a DataFrame
# and keeping the original column names from Before_imputation
After_imputation=pd.DataFrame(After_imputation,columns=Before_imputation.columns)
After_imputation
```
