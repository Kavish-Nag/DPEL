```python
# Create a grouped bar chart
import numpy as np
import matplotlib.pyplot as plt
# Each tuple represents a group of values for different categories
data = ([30, 25, 50, 20],
        [40, 23, 51, 17],
        [35, 22, 45, 19])
X= np.arange(4)
fig = plt.figure()
# Add axes to the figure: [left, bottom, width, height] in relative (0–1) coordinates
ax = fig.add_axes([0,0,1,1])
# Plot bars for first dataset (blue), at original X positions
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25) #shifted right by 0.25 from original
ax,bar(X + 0.50, data[2], color = 'r', width = 0.25) #shifted right by 0.5 from original
plt.show()
```

```python
# Create a stacked bar chart
import numpy as np
import matplotlib.pyplot as plt
N = 5 #Number of groups
menMeanscore = (20, 35, 30, 35, 27)
womenMeanscore = (25, 32, 34, 20, 25)
ind = np.arange(N)    #the x locations for the groups
width = 0.35
fig = plt.figure()
#Add axes to the figure: [left, bottom, width, height] in relative (0–1) coordinates
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, menMeanscore, width, color='r')
ax.bar(ind, womenMeanscore, width, bottom=menMeanscore, color = 'g')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5')) #Group names
ax.set_yticks(np.arange(0, 81, 10)) #Set Y-axis ticks from 0 to 80 in steps of 10
ax.legend(('Men', 'Women'))
plt.show()
```

```python
# Create a histogram
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1,1) #Figure and a set of subplots (1 row, 1 column, 1 plot)
a = np.array([22,87,5,43,9,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins=[0, 25, 50, 75, 100]) #bins-defines the intervals: [0–25), [25–50), [50–75), [75–100)
ax.set_title("Histogram of Result")
ax.set_xticks([0, 25, 50, 75, 100])
ax.set_xlabel('Marks')
ax.set_ylabel('Number of Students')
plt.show()
```
