#Filtering DataFrames

Filtering is about rows

```python
Filter = stats['InternetUsers']<2
Filter
stats[Filter]
```

```python
# This line creates a boolean filter based on the 'InternetUsers' column
# It checks if the value in 'InternetUsers' is less than 2 for each row
Filter = stats['InternetUsers']<2

# This line prints the boolean filter series
# It shows True for rows where the condition is met and False otherwise
print(Filter)

# This line applies the boolean filter to the stats DataFrame
# It returns a new DataFrame containing only the rows where the filter is True
stats[Filter]
```

```python
#Filter by birthrate
Filter2=stats['BirthRate']>40
stats[Filter2]
```

```python
#Above can be done in one step
stats[stats['BirthRate']>40]
```

```python
#Applying 2  filters
stats[Filter & Filter2]
```

```python
#Apply two filters using square bracket approach
stats[(stats.BirthRate>40)&(stats.InternetUsers<2)] #need parenthesis for order of operations
```
