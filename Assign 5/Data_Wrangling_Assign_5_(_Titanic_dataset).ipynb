```python
#fill missing values in 'Age' columns
age_mean=df['Age'].mean()
df['Age']=df['Age'].fillna(age_mean)
df
```

```python
df.isna().sum() #Shows number of null values for each column
```

```python
#Forward fill method
df['Embarked'].fillna(method='ffill',inplace=True)
df.isna().sum()
```

```python
#create new title column named title
df['Title']=df['Name'].str.extract('([A-Za-z]+)\.',expand=True)
df['Title'].value_counts()
```

```python
df.head() #Displays first 5 rows
```

```python
#replacing rare titles with more common ones
mapping={'Mlle':'Miss','Major':'Mr','Col':'Mr','Sir':'Mr','Don':'Mr','Mme':'Mrs','Jonkheer':'Mr','Lady':'Mrs','Capt':'Mr','Countess':'Mrs','Ms':'Miss','Dona':'Mrs'}
df.replace({'Title':mapping},inplace=True)
```

```python
# Returns a Series where:
# - The index is the unique titles
# - The values are the counts, sorted in descending order by default
df['Title'].value_counts()
```

# Reshaping data

```python
#Categorize data
#Map 'male'- 0 and 'female'- 1 in 'Gender' column
df['Gender']=df['Gender'].map({'male':0,'female':1})
df
```

```python
# Map 0 - 'No' and 1 - 'Yes' in 'Survived' column
df['Survived']=df['Survived'].map({0:'No',1:'Yes'})
df
```
