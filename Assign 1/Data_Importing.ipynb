```python
#importing libraries
import numpy as np
import pandas as pd
import pickle
```

```python
def load_csv(filepath):
    data = []
    col = []
    checkcol = False  # Used to identify and store the first line as column headers
    with open(filepath) as f:
        for val in f.readlines():
            val = val.replace("\n", "").split(',')  # Clean and split each line
            if not checkcol:
                col = val        # First line is treated as header
                checkcol = True
            else:
                data.append(val)  # Remaining lines are data rows

    # Create and return a DataFrame using collected headers and data
    return pd.DataFrame(data=data, columns=col)
```

```python
myData = load_csv('/content/100 Sales Records.csv') #calling function
print(myData.head()) #head() function prints first 5 rows
```

```python
# Numpy.genfrontxt()
# text values gets changed to NaN (Not a number) value
data= np.genfromtxt('/content/100 Sales Records.csv',delimiter=',')
pd.DataFrame(data)
```

```python
#Convert NaN into other datatypes
data= np.genfromtxt('/content/100 Sales Records.csv',delimiter=',', dtype=  None)
#changes to byte represtation, prints 'b' before text
pd.DataFrame(data)
```

```python
#Keeps data in same dtype, they don't stay in byte representation
data=np.genfromtxt('/content/100 Sales Records.csv',delimiter=',',dtype=None,names=True)
pd.DataFrame(data)
```

```python
#To convert bytes format
data=np.genfromtxt('/content/100 Sales Records.csv',delimiter=',',dtype=None,names=True, encoding = 'utf-8')
# encoding='utf-8' ensures proper decoding of text data.
pd.DataFrame(data)
```
