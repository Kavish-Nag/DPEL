```python
#importing libraries
import pandas as pd
import uuid #to generate unique ids
import random #to generate random numbers
from faker import Faker #case sensitive makes it different
```

```python
#Number of rows or users to create
num_users=100
```

**Add feature List**
- ID - unique string of characters to identify users
- Gender - string datatype of three choices
- Subscriber - a binary True/False choice of their subscription status
- Name - string datatype o first and last name
- Rating - integer type of 1 through 5 rating of something.

```python
# A list of 5 features
features=['ID','Gender','Subscriber','Name','Rating']
#creating a df for these features
df=pd.DataFrame(columns=features)
df
```

Generate unique identifers
UUID is a great library to generate unique IDs for each user because of its astronomically low chance of duplicating an ID.

```python
# Creating a list of random unique numbers using uuid4 for each user
# uuid4() generates a random unique ID, and .node gives the number part of it
df['ID']=[uuid.uuid4().node for i in range(num_users)]
df
```

```python
#checking if all IDs are unique
print(df['ID'].nunique()==num_users)
#True = all ids are unique
#False = There are duplicate values
```

```python
#Generating values for genders
genders=['Male','Female','na']
df['Gender']=random.choices(genders,weights=(45,45,10),k=num_users)
#weights here means the ratio in which the data must be generated ( it should be mentioned in %)
#like here, 45 males, 45 females , 10 na
df
```

```python
df.Gender.value_counts() #counts the number of geneder values for each gender
```

```python
#Generating subscriber values
choice = [True,False]
df['Subscriber']=random.choices(choice,k=num_users) #picks a random choice from the given list of choice
df
```

```python
df.Subscriber.value_counts() #counts total values present
```

```python
#Generating fake names using faker library
faker=Faker(locale="FR_FR") #Call to faker library
def name_gen(gender):
  if gender=='Male':
    return faker.name_male()
  elif gender=='Female':
    return faker.name_female()
  else:
    return faker.name()

#generating names for each user
df['Name'] = [name_gen(i) for i in df['Gender']]
df
```

```python
ratings = [1,2,3,4,5] #different ratings available
#Weighted ratings with a skew towards the ends
df['Rating']=random.choices(ratings,weights=(5,10,10,10,5),k=num_users)
df
```
